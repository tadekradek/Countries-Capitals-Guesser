```python
# create and empty column in world called 'capital'
world['capital'] = None

# create a dictionary from capital dataframe where keys are Państwo and values are Stolica
capital_dict = dict(zip(capital.Państwo, capital.Stolica))

# for each row in world, if name_pl is in capital_dict, assign capital_dict[name_pl] to respective world['capital']
for i in range(len(world)):
    if world['name_pl'][i] in capital_dict:
        world['capital'][i] = capital_dict[world['name_pl'][i]]
    else:
        world['capital'][i] = None

world
```
